```python
from datasets import load_dataset

dataset = load_dataset("librispeech_asr",
                       split="train.clean.100",
                       streaming=True,
                       trust_remote_code=True)
```

```python
dataset_head = dataset.take(5)
list(dataset_head)

```

```python
from IPython.display import Audio as IPythonAudio
example = list(dataset_head)[2]
print(f"Text of the following audio: {example['text']}")
IPythonAudio(example["audio"]["array"],
             rate=example["audio"]["sampling_rate"])
```

```python
from transformers import pipeline
asr = pipeline("automatic-speech-recognition", model="openai/whisper-small", device=0)
```

```python
model_acceptable_rate = asr.feature_extractor.sampling_rate
data_rate = example['audio']['sampling_rate']

print(f'model required sampling rate vs. data sampling rate: {model_acceptable_rate} vs. {data_rate}')
```

```python
result = asr(example["audio"]["array"])
print('Result: ', result['text'])
print('Source: ', example['text'])
```
